````python
from matplotlib import pyplot as plt
````

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
````

__Exercise 2 (A simple mathematical plot)__:
1. Use NumPy to generate $x$ and $y$ data suitable for plotting the graph of the function:
$y=e^{-x/10}sin(x)$

for values of $x$ in the range 0 to 50.

1. Customise your plot so that:

    1. The x and y axes have labels of fontsize of 15
    1. The x and y axes labels are in the color, darkorange.
    1. There is a grid with an opacity of 25%.
    1. Indicate the equation of the function in the title with a font size of 20.
    Hint: Look at Writing mathematical expressions in Matplotlib.
    1. Change the $y$ limit (using `ylim`) to [-1,1]
In the end, your plot should look like the one shown alongside.


````python
import numpy as np
````

````python
# Define x using np.linspace()
x = np.linspace(0, 50, 100)

# Calculate y values based on x
y = np.exp(-x / 10) * np.sin(x)

# Plotting the data
plt.plot(y, label='e^{-x/10}sin(x)', color='steelblue', linestyle='solid')
plt.ylim(-1, 1)
plt.xlabel('x', color='darkorange', fontsize=15)
plt.ylabel('y', color='darkorange', fontsize=15, rotation=90)
plt.title('$y = e^{-x/10}sin(x)$', fontsize=20)
plt.grid(alpha=0.25)

#Loose ends:
#Hm for some reason I can't rotate the y label and I can't 
#get the same smoothness within a range of 0 to 50. 
````
