# 1 A simple plot

## 1.1 Letâ€™s look at some code.

Shown below is some code and the resulting plot.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed') #plots the axis according to the data
plt.xlabel('x-values') #produces x-axis label
plt.ylabel('y-values') #produces y-axis label
plt.title('X vs Y') #produces title
plt.grid(alpha=.25) #produces gridlines
plt.legend(loc='upper left') #produces a legen in the upper left corner of plot
````

A simple way to do this is by commenting out some lines to turn off their contributions or changing some parameters (e.g., `loc` to `bottom left`) and numbers (e.g., `alpha` to `.8`). It will help you learn faster if you try to predict what will happen before running the code.

### Things to note

You can use the following abbreviations if you like:
| Long form | Abbreviation |
|:---------:|:------------:|
|   `color`   |       `c`      |
| `linestyle` |      `ls`      |
| `linewidth` |      `lw`      |

so, both the following lines produce the same result:



````python
# You can plot with the full length terms:
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2, label='Y values')

plt.xlabel('x-values') 
plt.ylabel('y-values') 
plt.title('X vs Y')
plt.grid(alpha=.25) 
plt.legend(loc='upper left')
````

````python
# Or, the same can be achieved with abbreviated forms c, ls & lw
plt.plot(x, y_1, c='red', ls='dashed', lw=2, label='Y values')

plt.xlabel('x-values') 
plt.ylabel('y-values') 
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

# 2 Better with NumPy

Often, it is easier to use NumPy arrays instead of Python lists. So, letâ€™s first convert the Python lists into NumPy arrays and then redo the plot in the previous step.



````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

import numpy as np

# convert the python lists into numpy arrays
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

## 2.1 Adding mathematical functions
Now we get to why using numpy arrays are better!
So, one of the __advantages of NumPy arrays__ is that __they allow us to generate data-related mathematical functions easily__. Letâ€™s reuse our previous code to plot 
$x^2$ and $sin(x)$

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')           


plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

Ah but alas, our plot does not look good because 
$sin(x)$ lies between $\pm1$, but $x^2$
 has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis.

## We need another axis!

Matplotlib offers a variety of ways to have multiple axes. The simplest way is to __have another y-axis__ that shares the same x-axis. We can _use the command `twinx()` for this._

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

### Things to note

- We now have two `legend()` calls, one for each axis.

- our plot still does not look good because we have only a few points. Letâ€™s use `np.linspace()` to fix this with:

````python
#Use np.linespace rather than x = np.array([0, 1, 2, 3, 4, 5, 6]) which plots makes 6 points

x = np.linspace(0, 6, 100) # specifices to create 100 equally spaced points between 0 and 6. 
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

# 3 Saving to disc

If you want to use your plot in a report or presentation, you must first __save it to disk__. Luckily, Matplotlib makes it astonishingly easy to export plots into many formats (PDF, JPEG, PNG, BMP, etc.) and at different resolutions.

- We need to use the function `savefig()` for this.
- We specify the format with the extension (e.g., filename.pdf) of the file name,
- We specify resolution by use dots-per-inch (`dpi`).

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.savefig('simple-plot.png', dpi=150)
#Rmbr to place your plt.savefig() on the very last line so 
#you're saving all aspects of your plot
````

Thatâ€™s all! Yes, itâ€™s that easy!

When you run this code, you will find the file saved in the same directory (folder) as the one in your notebook lives in. If you want it saved elsewhere, specify the path in more detail. For example:

- Windows `plt.savefig('C://Desktop/simple-plot.png', dpi=150)`
- macOS `plt.savefig('~/Desktop/simple-plot.png', dpi=150)`

# 4 A real example: Global Warming

## 4.1 Plotting data from files

Plotting data stored in a file (e.g., spreadsheet, text file, database) is a routine task for a scientist. In fact, the first thing you should do with any data is to look at it with a simple plot.

For the rest of this section, I will use the Earthâ€™s land temperature data from the Berkeley Earth website. Please visit the site (Global Warming 
 Data Overview) and download the average temperature data for Daily Land. The original name of the file should be `Complete_TAVG_daily.txt`

It looks like this:

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

# Load data from the file 'Complete_TAVG_daily.txt', skipping the first 24 rows
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

# Extract date and anomaly columns from the loaded data
date = data[:, 0]       # Since date column is the first column
anomaly = data[:, -1]   # reverese indexing since the anomaly column is the last column

# Plot date against anomaly
plt.plot(date, anomaly, alpha=0.5)  # Alpha parameter controls transparency of the plot

plt.ylim([-8, 8]) # Set y-axis limits
plt.xlabel('Year') # Label x-axis

plt.ylabel('Temperature ($^\circ$C)') # Label y-axis
plt.title('Global Average Daily Temperature') # Set plot title
plt.grid(alpha=0.25) # Display gridlines with specified transparency 
````

I have used a _small alpha value to soften the colour_ of the plot and made the plot range symmetrical in the 
$y$ direction.

Letâ€™s now add a horizontal line at the zero value to highlight the trend shown by the data.
The `hlines()` function needs a 
$y$-value and starting and ending values for $x$



````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

# Load data from the file 'Complete_TAVG_daily.txt', skipping the first 24 rows
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

# Extract date and anomaly columns from the loaded data
date = data[:, 0]       # Since date column is the first column
anomaly = data[:, -1]   # reverese indexing since the anomaly column is the last column

# Plot date against anomaly
plt.plot(date, anomaly, alpha=0.5)  # Alpha parameter controls transparency of the plot

plt.ylim([-8, 8]) # Set y-axis limits
plt.xlabel('Year') # Label x-axis

plt.ylabel('Temperature ($^\circ$C)') # Label y-axis
plt.title('Global Average Daily Temperature') # Set plot title
plt.grid(alpha=0.25) # Display gridlines with specified transparency 
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') # adds horizontal line
````
